```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV 
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.impute import KNNImputer
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("world_bank.csv")
```

```python
# Cell 3: Trim to years 2002–2020
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year.between(2002, 2020)].copy()

#Sanity Check
print("Data covers:", df['date'].dt.year.min(), "→", df['date'].dt.year.max())
```

```python
# Cell 4: Drop discontinued or weak features
columns_to_drop = [
    'doing_business',
    'time_to_get_operation_license',
    'multidimensional_poverty_headcount_ratio%',
    'human_capital_index',
    'risk_premium_on_lending',
    'electric_power_consumption',
    'logistic_performance_index',
    'statistical_performance_indicators',
    'avg_precipitation',
    'goverment_effectiveness_std',
    'regulatory_quality_std',
    'voice_and_accountability_std',
    'control_of_corruption_std',
    'rule_of_law_std',
    'political_stability_std',
]
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Cell 5: Remove countries ≥ 90% missing across all remaining columns
pct_missing = df.isna().groupby(df['country']).mean().mean(axis=1)
to_drop = pct_missing[pct_missing >= 0.9].index
df = df[~df['country'].isin(to_drop)].copy()
```

```python
# Cell 6: Remove any country that never reports Gini
has_gini = df.groupby('country')['gini_index'].transform('count') > 0
df = df[has_gini].copy()
```

```python
# Cell 7: Year-by-Year Missing Data (%) for Top-20 Most Incomplete Countries
# — build year column
df['year'] = df['date'].dt.year

# — compute per-row % missing across all non-ID cols
data_cols = df.columns.difference(['country', 'date', 'year'])
df['missing_pct'] = df[data_cols].isna().mean(axis=1) * 100

# — pivot to country × year, averaging if multiple records
heatmap_df = df.pivot_table(
    index='country',
    columns='year',
    values='missing_pct',
    aggfunc='mean'
)

# — pick top 20 countries by overall missingness
top20 = heatmap_df.mean(axis=1).sort_values(ascending=False).head(20).index
heatmap_df = heatmap_df.loc[top20]

# — plot
plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_df,
    cmap='Reds',
    annot=True,
    fmt='.1f',
    linewidths=.4,
    linecolor='white',
    cbar_kws={'label': '% Missing'}
)
plt.title('Year-by-Year Missing Data (%)  •  Top-20 Most Incomplete Countries')
plt.xlabel('Year')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Cell 8: Drop the helper column and save final CSV
df.drop(columns=['missing_pct'], inplace=True)
```

```python
# Cell 9: we finally output the cleaned csv
df.to_csv('cleaned.csv', index=False)
```

```python
# Cell 10: Country-mean imputation with global fallback
impute_cols = df.columns.difference(['country', 'date', 'year', 'gini_index'])
df_mean = df.copy()

# 1️⃣ Fill per-country mean; 2️⃣ fallback to global mean
df_mean[impute_cols] = df_mean.groupby('country')[impute_cols] \
                              .transform(lambda g: g.fillna(g.mean()))
df_mean[impute_cols] = df_mean[impute_cols].fillna(df_mean[impute_cols].mean())

# 3️⃣ Compute 5-year rolling mean & std
df_mean['date'] = pd.to_datetime(df_mean['date'])
df_mean.sort_values(['country','date'], inplace=True)
rolled = df_mean.groupby('country')[impute_cols].rolling(5, min_periods=5)
df_mean[[f'{c}_rolling_mean' for c in impute_cols]] = rolled.mean().reset_index(level=0, drop=True)
df_mean[[f'{c}_rolling_std'  for c in impute_cols]] = rolled.std().reset_index(level=0, drop=True)

# 4️⃣ Drop pre-2007 rows and save
df_mean = df_mean[df_mean['date'].dt.year >= 2007]
df_mean.to_csv('cleaned_mean.csv', index=False)
```

```python
# Cell 11: KNN imputation
# !! Note the no. of neigbours is completely arbitrary at k = 100
df_knn = df.copy()
df_knn[impute_cols] = KNNImputer(n_neighbors=100).fit_transform(df_knn[impute_cols])
df_knn['date'] = pd.to_datetime(df_knn['date'])
df_knn.sort_values(['country','date'], inplace=True)
rolled = df_knn.groupby('country')[impute_cols].rolling(5, min_periods=5)
df_knn[[f'{c}_rolling_mean' for c in impute_cols]] = rolled.mean().reset_index(level=0, drop=True)
df_knn[[f'{c}_rolling_std'  for c in impute_cols]] = rolled.std().reset_index(level=0, drop=True)
df_knn = df_knn[df_knn['date'].dt.year >= 2007]
df_knn.to_csv('cleaned_knn.csv', index=False)
```

```python
# Cell 12: MICE imputation
df_mice = df.copy()
df_mice[impute_cols] = IterativeImputer(random_state=42).fit_transform(df_mice[impute_cols])
df_mice['date'] = pd.to_datetime(df_mice['date'])
df_mice.sort_values(['country','date'], inplace=True)
rolled = df_mice.groupby('country')[impute_cols].rolling(5, min_periods=5)
df_mice[[f'{c}_rolling_mean' for c in impute_cols]] = rolled.mean().reset_index(level=0, drop=True)
df_mice[[f'{c}_rolling_std'  for c in impute_cols]] = rolled.std().reset_index(level=0, drop=True)
df_mice = df_mice[df_mice['date'].dt.year >= 2007]
df_mice.to_csv('cleaned_mice.csv', index=False)
```

```python
# Cell 13: Faster MissForest imputation
df_forest = df.copy()
df_forest[impute_cols] = IterativeImputer(
    estimator=RandomForestRegressor(n_estimators=10, max_depth=5, n_jobs=-1, random_state=42),
    max_iter=5,
    random_state=42
).fit_transform(df_forest[impute_cols])
df_forest['date'] = pd.to_datetime(df_forest['date'])
df_forest.sort_values(['country','date'], inplace=True)
rolled = df_forest.groupby('country')[impute_cols].rolling(5, min_periods=5)
df_forest[[f'{c}_rolling_mean' for c in impute_cols]] = rolled.mean().reset_index(level=0, drop=True)
df_forest[[f'{c}_rolling_std'  for c in impute_cols]] = rolled.std().reset_index(level=0, drop=True)
df_forest = df_forest[df_forest['date'].dt.year >= 2007]
df_forest.to_csv('cleaned_forest.csv', index=False)
```

```python
# Cell 14: Check for remaining NaNs in each cleaned CSV (exclude ID & target cols)
files = {
    'Mean':   'cleaned_mean.csv',
    'KNN':    'cleaned_knn.csv',
    'MICE':   'cleaned_mice.csv',
    'Forest': 'cleaned_forest.csv'
}
exclude = ['gini_index']

for name, path in files.items():
    df_chk = pd.read_csv(path)
    na_counts = df_chk.drop(columns=exclude).isna().sum()
    na_counts = na_counts[na_counts > 0]
    print(f"\n{name} — remaining NaNs:")
    if na_counts.empty:
        print("  None")
    else:
        print(na_counts)
```

```python
# Cell 15: Country-level 80/20 train-test split without country overlap
files = {
    'mean':   'cleaned_mean.csv',
    'knn':    'cleaned_knn.csv',
    'mice':   'cleaned_mice.csv',
    'forest': 'cleaned_forest.csv'
}

for name, path in files.items():
    df = pd.read_csv(path)
    countries = df['country'].unique()
    train_ctrs, test_ctrs = train_test_split(countries, test_size=0.2, random_state=42)
    df_train = df[df['country'].isin(train_ctrs)].reset_index(drop=True)
    df_test  = df[df['country'].isin(test_ctrs)].reset_index(drop=True)
    df_train.to_csv(f'{name}_train.csv', index=False)
    df_test .to_csv(f'{name}_test.csv',  index=False)
```
