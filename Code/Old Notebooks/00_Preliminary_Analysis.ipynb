```python
# ──────────────────────────────────────────────────────────────
# Block 4  ▸  Feature‑level missing‑data audit
# -----------------------------------------------------------------
#  Goal: identify which columns still contain NaNs and how severe
#        the missingness is, then visualise the worst offenders.
# -----------------------------------------------------------------

import matplotlib.pyplot as plt

# 1️⃣  Calculate % missing for every column -------------------------------
missing_pct = (df.isna().mean() * 100).sort_values(ascending=False)

# 2️⃣  Retain only columns that have at least one missing value ----------
missing_pct = missing_pct[missing_pct > 0]

print("Columns with missing values (sorted):\n")
print(missing_pct.round(2))

# 3️⃣  Bar‑plot of missingness -------------------------------------------
plt.figure(figsize=(12, 6))
missing_pct.plot(kind='bar')
plt.ylabel('% Missing')
plt.title('Missing Data Percentage by Feature')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ──────────────────────────────────────────────────────────────
# Block 5  ▸  Drop ultra‑sparse or conceptually irrelevant columns
# -----------------------------------------------------------------
#  Rationale for each group of drops is noted inline so future readers
#  understand *why* these features were removed.
#  Note that these are subject to change as our group's understanding of the dataset evolves.
# -----------------------------------------------------------------

columns_to_drop = [
    # ── Discontinued / sparsely‑reported indices (≳ 70 % NaNs) ────────────
    'doing_business',                       # 95 % NaN; WB index retired
    'time_to_get_operation_license',        # 94 % NaN; sub‑indicator of Doing Business
    'multidimensional_poverty_headcount_ratio%',  # 88 % NaN; overlaps with Gini conceptually
    'human_capital_index',                  # 82 % NaN; composite metric, reported for few years
    'risk_premium_on_lending',              # 76 % NaN; niche financial risk measure

    # ── Mid‑sparse features & weak conceptual link to inequality ─────────
    'electric_power_consumption',           # 60 % NaN; energy use, poor coverage
    'logistic_performance_index',           # 52 % NaN; trade‑logistics score
    'statistical_performance_indicators',   # 48 % NaN; capacity metric unrelated to income distribution
    'avg_precipitation',                    # 50 % NaN; climate variable, weak causal tie to Gini

    # ── Governance *uncertainty* columns  (we keep point estimates) ──────
    'goverment_effectiveness_std',
    'regulatory_quality_std',
    'voice_and_accountability_std',
    'control_of_corruption_std',
    'rule_of_law_std',
    'political_stability_std',
]

# ---- Execute the drop ----------------------------------------------------
df = df.drop(columns=columns_to_drop, errors='ignore')

print(f"Dropped {len(columns_to_drop)} sparse / non‑essential columns.")
```

```python
# ──────────────────────────────────────────────────────────────
# Block 6  ▸  Re‑audit feature‑level missingness after drops
# -----------------------------------------------------------------
#  Purpose : confirm which columns still contain NaNs and inspect
#            their severity now that we have removed the worst offenders.
# -----------------------------------------------------------------

import matplotlib.pyplot as plt

# 1️⃣  Re‑compute % missing for every remaining column
missing_pct = (
    df.isna().mean()            # fraction of rows that are NaN
      .mul(100)                 # → %
      .loc[lambda s: s > 0]     # keep only columns with any missingness
      .sort_values(ascending=False)
)

print("\nRemaining columns with missing values:\n")
print(missing_pct.round(2))

# 2️⃣  Visualise the remaining missingness
plt.figure(figsize=(12, 6))
missing_pct.plot(kind='bar')
plt.ylabel('% Missing')
plt.title('Missing Data Percentage by Feature (Post‑Pruning)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ──────────────────────────────────────────────────────────────
# Block 7  ▸  Country‑level missing‑data dashboard
# -----------------------------------------------------------------
#  Goal       : create a tidy DataFrame that tells us, for every
#               country,
#                 • its overall % of missing cells, and
#                 • the three features with the highest missingness.
#  Output     : `country_missing_df`
# -----------------------------------------------------------------
#  Assumptions: • df   already cleaned / trimmed to 2002‑2020
#               • 'country' column present
# -----------------------------------------------------------------

# 1️⃣  Overall % missing per country ----------------------------------------
country_counts = df.groupby('country').size()            # rows per country
total_cells    = country_counts * df.shape[1]            # rows × columns
null_counts    = (
    df.isna()
      .groupby(df['country'])
      .sum()                                            # NaNs per column
      .sum(axis=1)                                      #   … summed over cols
)
missing_pct_country = (null_counts / total_cells) * 100  # % missing overall

# 2️⃣  % missing *per feature* within each country -------------------------
missing_pct_by_country = (
    df.groupby('country')
      .apply(lambda g: g.isna().mean() * 100)            # % missing col‑wise
)

# 3️⃣  Assemble summary DataFrame ------------------------------------------
rows = []
for ctry, overall in missing_pct_country.sort_values(ascending=False).items():
    # top‑3 features driving missingness for this country
    top3 = (
        missing_pct_by_country.loc[ctry]
                              .sort_values(ascending=False)
                              .head(3)
                              .index.tolist()
    )
    rows.append(
        dict(country=ctry,
             overall_missing_pct=overall,
             top1_feature=top3[0],
             top2_feature=top3[1],
             top3_feature=top3[2])
    )

country_missing_df = pd.DataFrame(rows).reset_index(drop=True)

# 4️⃣  Peek at the ten worst offenders --------------------------------------
country_missing_df.head(10)
```

```python
# ──────────────────────────────────────────────────────────────
# Block 8  ▸  Heatmap of missingness  |  Top‑20 most incomplete countries
# -----------------------------------------------------------------
#  Purpose : visualise, for the 20 countries with the highest overall
#            missingness, how the %‑missing varies year‑by‑year (2002‑2020).
# -----------------------------------------------------------------
#  Requires : • df   – cleaned DataFrame, filtered to 2002‑2020 window
#             • sns  – seaborn  (import seaborn as sns)
#             • plt  – matplotlib.pyplot
# -----------------------------------------------------------------

# 1️⃣  Make sure we have a 'year' helper column
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year

# 2️⃣  Identify the 20 “worst” countries by overall %‑missing -------------
overall_missing_pct = (
    df.isna()
      .groupby(df['country']).sum().sum(axis=1)              # total NaNs / country
      / (df.groupby('country').size() * df.shape[1]) * 100   # → %
)

top20_countries = overall_missing_pct.nlargest(20).index.tolist()

# 3️⃣  Compute %‑missing for each (country, year) combination ------------
missing_by_ctry_year = (
    df.groupby(['country', 'year'])
      .apply(lambda g: g.isna().mean().mean() * 100)          # mean over rows & cols
      .reset_index(name='missing_pct')
)

# 4️⃣  Pivot to Country × Year matrix, newest years on the right ---------
heatmap_df = (
    missing_by_ctry_year
      .pivot(index='country', columns='year', values='missing_pct')
      .reindex(top20_countries)                     # maintain “worst‑first” order
      .sort_index(axis=1, ascending=False)          # years: newest → oldest
)

# 5️⃣  Plot heatmap -------------------------------------------------------
plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_df,
    cmap='Reds',
    annot=True,
    fmt='.1f',
    linewidths=.4,
    linecolor='white',
    cbar_kws={'label': '% Missing'}
)
plt.title('Year‑by‑Year Missing Data (%)  •  Top‑20 Most Incomplete Countries')
plt.xlabel('Year')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# ──────────────────────────────────────────────────────────────
# Block 9  ▸  Remove countries that are effectively “all missing”
# -----------------------------------------------------------------
#  Any country with ≥ 90 % of its cells missing across 2002‑2020 is
#  excluded from further analysis—they contribute little signal
#  and complicate imputation / modelling.
# -----------------------------------------------------------------

# 1️⃣  Overall %‑missing per country ----------------------------------------
overall_missing_pct = (
    df.isna()                                   # boolean mask
      .groupby(df['country']).sum().sum(axis=1) # total NaNs per country
      / (df.groupby('country').size()           # rows × columns
         * df.shape[1]) * 100
)

# 2️⃣  Identify “ghost” countries (≥ 90 % missing) -------------------------
to_drop = overall_missing_pct[overall_missing_pct >= 90].index.tolist()
print(f"Dropping {len(to_drop)} near‑empty countries:\n", to_drop)

# 3️⃣  Remove them from the DataFrame --------------------------------------
df = (
    df.loc[~df['country'].isin(to_drop)]
      .reset_index(drop=True)
)

print("Remaining countries:", df['country'].nunique())
```

```python
# ──────────────────────────────────────────────────────────────
# Block 14  ▸  Country‑year matrix of Gini availability  (2002‑2020)
# -----------------------------------------------------------------
#  Creates a Country × Year table where each cell is
#      1  →  Gini reported
#      0  →  Gini missing
#  Then adds a "missing_pct" column so we can rank countries by
#  how often they *fail* to report a Gini in the 2002‑2020 window.
# -----------------------------------------------------------------
import pandas as pd

# 1️⃣  Ensure a clean Year column ------------------------------------------
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year

# 2️⃣  Keep only rows within our analysis window (2002‑2020) --------------
df = df[df['year'].between(2002, 2020)].copy()

# 3️⃣  Binary flag: 1 if Gini present, 0 otherwise -------------------------
df['gini_present'] = df['gini_index'].notna().astype(int)

# 4️⃣  Pivot to Country × Year matrix --------------------------------------
presence_pivot = df.pivot_table(
    index='country',
    columns='year',
    values='gini_present',
    aggfunc='max',     # in case a country has >1 row per year, treat “any 1” as present
    fill_value=0
)

# 5️⃣  For each country, compute % of years with *missing* Gini -----------
years = presence_pivot.columns.to_list()
presence_pivot['missing_pct'] = (
    (presence_pivot[years] == 0).sum(axis=1) / len(years) * 100
)

# 6️⃣  Rank countries by missing_pct  (highest → lowest) -------------------
presence_pivot = presence_pivot.sort_values('missing_pct', ascending=False)

# 7️⃣  Reset index so 'country' becomes a column for easy viewing ----------
result_df = presence_pivot.reset_index()

# 8️⃣  Quick peek at the 20 worst (most missing) countries -----------------
result_df.head(20)
```

```python
# ──────────────────────────────────────────────────────────────
# Block 15  ▸  Align features (year t) with the *next‑survey* Gini (year t + 1)
# -----------------------------------------------------------------
#  Goal : build `model_df` where each row contains
#          • macro features for a given (country, year = t)
#          • target column `gini_next`  →  the Gini reported at year t + 1
#  We drop rows that do not have a *following‐year* Gini so that every
#  observation is a complete (X, y) pair ready for modelling.
# -----------------------------------------------------------------

# 1️⃣  Ensure a clean 'year' helper column ----------------------------------
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year

# 2️⃣  Sort by country & year, then create the shifted target --------------
df = df.sort_values(['country', 'year'])

# For each country, shift Gini *up* by one row: year t row gets Gini(t + 1)
df['gini_next'] = df.groupby('country')['gini_index'].shift(-1)

# 3️⃣  Keep only rows that *do* have a next‑year Gini ----------------------
model_df = (
    df.loc[df['gini_next'].notna()]   # drop rows where gini_next is NaN
      .reset_index(drop=True)
)

# 4️⃣  Quick sanity‑check ---------------------------------------------------
print("Modelling table shape:", model_df.shape)
model_df
```

```python
# ──────────────────────────────────────────────────────────────
# Block 16  ▸  Missing‑data audit for the *modelling* table (`model_df`)
# -----------------------------------------------------------------
#  Context : `model_df` keeps only those (country, year =t) rows that
#            have a valid `gini_next` at year =t + 1.
#  Task    : inspect which *predictor* columns still contain NaNs and
#            visualise their severity.
# -----------------------------------------------------------------
import matplotlib.pyplot as plt
import pandas as pd

# 1️⃣  Choose the set of feature columns (exclude target & helpers) ---------
feature_cols = [
    c for c in model_df.columns
    if c not in ('gini_index', 'gini_next', 'date', 'year')  # exclude helpers / targets
]

# 2️⃣  Compute % missing for each remaining feature ------------------------
missing_pct = (
    model_df[feature_cols]
      .isna()
      .mean() * 100
)

# Keep only columns that still have any NaNs
missing_pct = (
    missing_pct[missing_pct > 0]
      .sort_values(ascending=False)
)

print("\nMissingness in modelling features (sorted):\n")
print(missing_pct.round(2))

# 3️⃣  Bar‑plot of missingness --------------------------------------------
if not missing_pct.empty:
    plt.figure(figsize=(12, 6))
    missing_pct.plot(kind='bar')
    plt.ylabel('% Missing')
    plt.title('Missing Data Percentage by Feature  •  `model_df` (X variables)')
    plt.xticks(rotation=90, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("\n✅  No missing values in the features of `model_df`.")
```

# Lasso Regressor With Country as a Feature

```python
# ──────────────────────────────────────────────────────────────
# Block 17  ▸  Assemble feature matrix (X) and target vector (y)
# -----------------------------------------------------------------
#  X  : all macro predictors at year t
#  y  : Gini that will be observed at year t + 1   (column `gini_next`)
#
#  We drop:
#    • any column that is itself a target or helper (`gini_*`, dates, etc.)
#    • leakage flags we created earlier (`next_year`, `gini_present`)
# -----------------------------------------------------------------

cols_to_drop = [
    'gini_index',   # Gini at year t          (leaks information from same year)
    'gini_next',    # ← this *becomes* y, so remove from X
    'next_gini',    # alias, if still present
    'date', 'year', # time helpers, not used as features
    'next_year',    # survey‑timing helper  (would leak target year)
    'gini_present'  # binary flag of availability (also leaks)
]

# ── Build the design matrix X and target vector y ─────────────────────────
X = model_df.drop(columns=cols_to_drop, errors='ignore')
y = model_df['gini_next']

print("X shape :", X.shape)
print("y shape :", y.shape)
X.head()
```

```python
# ──────────────────────────────────────────────────────────────
# Block 18  ▸  One‑hot country dummies  ➜  row‑wise train/test split
# -----------------------------------------------------------------
#  ⚠️  This version **keeps country dummies** *and* uses a random
#      row‑level split.  We keep it here on purpose to illustrate how
#      “country leakage” can inflate test‑set scores.
#
#      • For a leakage‑free model, see Block 19 (group‑split).
# -----------------------------------------------------------------

from sklearn.model_selection import train_test_split
from sklearn.impute        import SimpleImputer
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 1️⃣  One‑hot‑encode 'country'  (drop_first=True avoids dummy trap) -------
X = pd.get_dummies(X, columns=['country'], drop_first=True)

# 2️⃣  Row‑wise 80 / 20 train‑test split  (leakage demo version) ----------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# 3️⃣  Identify numeric vs. dummy columns ---------------------------------
dummy_cols = [c for c in X_train.columns if c.startswith('country_')]
num_cols   = [c for c in X_train.columns if c not in dummy_cols]

# 4️⃣  Impute missing numeric values (mean) -------------------------------
imputer = SimpleImputer(strategy='mean')
X_train[num_cols] = imputer.fit_transform(X_train[num_cols])
X_test [num_cols] = imputer.transform(    X_test [num_cols])

# 5️⃣  Standardise numeric features (0‑mean, unit‑var) --------------------
scaler = StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test [num_cols] = scaler.transform(    X_test [num_cols])

# 6️⃣  Quick peek at the prepared design matrix ---------------------------
print("Design matrix shape:", X_train.shape)
X_train.head()
```

```python
# ──────────────────────────────────────────────────────────────
# Block 19  ▸  Lasso hyper‑parameter tuning (row‑wise split version)
# -----------------------------------------------------------------
#  This block keeps the “row‑wise” train/test split from Block 18.
#  Because country dummies are present *and* rows from each country
#  appear in both sets, the resulting scores will be *optimistic*.
#  We keep this run to illustrate the leakage effect.
# -----------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model  import Lasso
from sklearn.model_selection import GridSearchCV

# 1️⃣  Hyper‑parameter grid (α spanning 1e‑4 … 1e5 in log‑10 steps) -------
alpha_grid = np.logspace(-4, 5, num=9)   # [1e‑4, 1e‑3, …, 1e5]
param_grid = {'alpha': alpha_grid}

# 2️⃣  Base Lasso model -----------------------------------------------------
lasso_base = Lasso(max_iter=5000, random_state=42)

# 3️⃣  5‑fold CV grid‑search (row‑wise folds) ------------------------------
gscv = GridSearchCV(
    estimator=lasso_base,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    return_train_score=True,
    n_jobs=-1
)
gscv.fit(X_train, y_train)

print(f"\nBest α  : {gscv.best_params_['alpha']}")
print(f"Mean CV R² (validation): {gscv.best_score_:.3f}")

# 4️⃣  Visualise train vs. validation R² across α --------------------------
cv_res = pd.DataFrame({
    'log10_alpha': np.log10(alpha_grid),
    'train_R2'   : gscv.cv_results_['mean_train_score'],
    'valid_R2'   : gscv.cv_results_['mean_test_score']
}).set_index('log10_alpha')

ax = cv_res[['train_R2', 'valid_R2']].plot(marker='o')
ax.set_xlabel('log₁₀(α)')
ax.set_ylabel('Cross‑validated R²')
ax.set_title('Lasso  •  Train vs. Validation Performance (row‑wise CV)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

# 5️⃣  Inspect non‑zero coefficients ---------------------------------------
best_lasso = gscv.best_estimator_
coef_series = pd.Series(best_lasso.coef_, index=X_train.columns)
nonzero_coef = coef_series[coef_series != 0].sort_values()

print("\nNumber of non‑zero coefficients:", nonzero_coef.size)
print("\nTop 20 (by magnitude):")
print(nonzero_coef.abs().sort_values(ascending=False).head(20))
```

# Linear Regression without country as a feature and make sure that the same country is not in both test and train 

```python
# ──────────────────────────────────────────────────────────────
# Block 20  ▸  Leakage‑free train / test split  (by country groups)
# -----------------------------------------------------------------
#  Goal : keep *whole* countries in either the training or the test set
#         so that no country appears in both.  This prevents the model
#         from “memorising” a country‑specific intercept.
# -----------------------------------------------------------------
from sklearn.model_selection import GroupShuffleSplit

# 1️⃣  Feature matrix & target vector  --------------------------------------
cols_to_drop = [
    'gini_index',       # Gini at year t      (target variant, leakage)
    'gini_next',        # becomes y
    'next_gini',        # alias
    'date', 'year',     # helpers
    'next_year',        # survey‑timing helper (leakage)
    'gini_present'      # same as above
    # note: keep 'country' only for grouping; we’ll remove it from X later.
]

X = model_df.drop(columns=cols_to_drop, errors='ignore')
y = model_df['gini_next']
groups = model_df['country']        # grouping key

# 2️⃣  Grouped 80 / 20 split  ----------------------------------------------
gss = GroupShuffleSplit(test_size=0.20, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups=groups))

X_train, X_test = X.iloc[train_idx].copy(), X.iloc[test_idx].copy()
y_train, y_test = y.iloc[train_idx],           y.iloc[test_idx]

# 3️⃣  Remove the raw 'country' column from the feature set (prevents leakage)
if 'country' in X_train.columns:
    X_train.drop(columns='country', inplace=True)
    X_test .drop(columns='country', inplace=True)

# 4️⃣  Sanity‑check: ensure no country overlap -----------------------------
assert set(groups.iloc[train_idx]).isdisjoint(set(groups.iloc[test_idx]))

print(f"Train set : {len(X_train):,} rows  |  {groups.iloc[train_idx].nunique()} countries")
print(f"Test  set : {len(X_test):,} rows  |  {groups.iloc[test_idx].nunique()} countries")
```

```python
# ──────────────────────────────────────────────────────────────
# Block 21  ▸  Impute & scale numeric features  (group‑split version)
# -----------------------------------------------------------------
#  After the group‑aware split we:
#    • drop any residual 'country' column from X_train/X_test
#    • mean‑impute remaining NaNs
#    • standardise each numeric feature (0‑mean, unit‑var)
# -----------------------------------------------------------------
from sklearn.impute        import SimpleImputer
from sklearn.preprocessing import StandardScaler

# 1️⃣  Make doubly sure the raw 'country' column is gone -------------------
X_train = X_train.drop(columns='country', errors='ignore')
X_test  = X_test .drop(columns='country', errors='ignore')

# 2️⃣  Mean‑imputation of missing numeric values --------------------------
imputer = SimpleImputer(strategy='mean')
X_train.iloc[:] = imputer.fit_transform(X_train)
X_test .iloc[:] = imputer.transform(   X_test )

# 3️⃣  Standardise  (keep DataFrame output) -------------------------------
scaler = StandardScaler().set_output(transform='pandas')
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(   X_test )

# 4️⃣  Quick confirmation --------------------------------------------------
print("Design matrix shapes  →  train:", X_train.shape, "|  test:", X_test.shape)
X_train.head()
```

```python
# ──────────────────────────────────────────────────────────────
# Block 22  ▸  LinearRegression vs. “mean” (bias‑only) regressor
#              ‑‑‑ nicer summary when some R² < 0
# -----------------------------------------------------------------
from sklearn.linear_model import LinearRegression
from sklearn.dummy       import DummyRegressor
from sklearn.metrics      import r2_score, mean_squared_error
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣  Fit the two models ---------------------------------------------------
lr_model   = LinearRegression()
bias_model = DummyRegressor(strategy='mean')

lr_model.fit(X_train, y_train)
bias_model.fit(X_train, y_train)

# 2️⃣  Predict on both splits ----------------------------------------------
y_pred = {
    'LR train'  : lr_model.predict(X_train),
    'LR test'   : lr_model.predict(X_test),
    'Bias train': bias_model.predict(X_train),
    'Bias test' : bias_model.predict(X_test)
}

mse_scores = pd.Series({
    name: mean_squared_error(y_train if 'train' in name else y_test, pred)
    for name, pred in y_pred.items()
}).round(2)
styled = (
    mse_scores.to_frame('MSE')
             .style
             .background_gradient(cmap='RdYlGn', vmin=-0.1, vmax=0.7)
             .format('{:.3f}')
)
display(styled)

# 3️⃣  Compute R² -----------------------------------------------------------
r2_scores = pd.Series({
    name: r2_score(y_train if 'train' in name else y_test, pred)
    for name, pred in y_pred.items()
}).round(3)

# 4️⃣  Nice tabular display -------------------------------------------------
styled = (
    r2_scores.to_frame('R²')
             .style
             .background_gradient(cmap='RdYlGn', vmin=-0.1, vmax=0.7)
             .format('{:.3f}')
)
display(styled)

# 5️⃣  Horizontal bar chart centred at 0 -----------------------------------
plt.figure(figsize=(7, 4))
colors = ['#b0b0b0' if 'Bias' in idx else '#1f77b4' for idx in r2_scores.index]
r2_scores.plot(kind='barh', color=colors, edgecolor='k')
plt.axvline(0, color='k', linewidth=0.8)         # centre line
plt.xlim(min(-0.1, r2_scores.min()*1.1), r2_scores.max()*1.1)
plt.xlabel('R²')
plt.title('Grouped‑Country Split  •  Model Comparison')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

# 6️⃣  Top coefficients of the linear model --------------------------------
coef_series = pd.Series(lr_model.coef_, index=X_train.columns)
top_coef = (
    coef_series.reindex(coef_series.abs().sort_values(ascending=False).index)
               .head(15)
               .round(3)
)

print("\nTop 15 LinearRegression coefficients (sorted by |weight|):")
display(top_coef.to_frame('β'))
```
