```python
# ──────────────────────────────────────────────────────────────
# Block 1  ▸  Load model and demo data
# -----------------------------------------------------------------
import pandas as pd
import joblib
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error

# Load the best‐performing model (exported earlier)
best_model   = joblib.load('best_gbr_model.pkl')

# Load the demo feature set, target, and country labels
X2_demo      = pd.read_csv('X2_demo.csv')
y_demo       = pd.read_csv('y_demo.csv')['gini_next']
groups_demo  = pd.read_csv('groups_demo.csv')['country']
```

# Take a guess at its Gini-Index (USA: 42%)

```python
# ──────────────────────────────────────────────────────────────
# Block 2  ▸  Evaluate for any custom country
# -----------------------------------------------------------------
country = 'NAME HERE'   # replace with desired name
mask    = (groups_demo == country)

X_sub   = X2_demo[mask]
y_sub   = y_demo[mask]
y_pred  = best_model.predict(X_sub)

print(f"\n--- Results for {country} ---")
pd.DataFrame({
    'actual'    : y_sub.values,
    'predicted' : y_pred.round(1),
    '% rel_error' : (100*(y_pred - y_sub.values)/y_sub.values).round(1)
}, index=y_sub.index)

```

```python
# ──────────────────────────────────────────────────────────────
# Block 3  ▸  Overall performance + per-country mean error bar plot
# -----------------------------------------------------------------
from sklearn.metrics import r2_score
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ Compute overall R² on the demo set
y_pred_all = best_model.predict(X2_demo)
overall_r2 = r2_score(y_demo, y_pred_all)
print(f"\nOverall Test R²: {overall_r2:.3f}")

# 2️⃣ Compute mean signed error per country (positive → over-prediction)
df_dev = pd.DataFrame({
    'country': groups_demo,
    'error': 100*(y_pred_all - y_demo)/y_demo
})
mean_error = df_dev.groupby('country')['error'] \
                   .mean() \
                   .sort_values()

# 3️⃣ Bar plot of mean error by country
plt.figure(figsize=(12,6))
mean_error.plot(kind='bar', color='steelblue')
plt.axhline(0, color='black', linewidth=1)            # zero-error line
plt.ylabel('% Relative Error')
plt.title('Per-Country % Relative Error\n(positive → over-predicted)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ──────────────────────────────────────────────────────────────
# Block 4  ▸  Evaluate for Suriname
# -----------------------------------------------------------------
country = 'Suriname'
mask    = (groups_demo == country)

X_sub   = X2_demo[mask]
y_sub   = y_demo[mask]
y_pred  = best_model.predict(X_sub)

results_su = pd.DataFrame({
    'actual'    : y_sub.values,
    'predicted' : y_pred,
    'abs_error' : -(y_pred - y_sub.values)
}, index=y_sub.index)

print(f"\n--- Results for {country} ---")
results_su
```
